# *** EDA insights *** 

## Weight Overview

```python
# Basic stats: min, max, mean, std

print("Weight Statistics:")
print(df[['WeightKg','BMI','Fat']].describe())
```

```python
# Histogram

df[['WeightKg','BMI','Fat']].hist(bins=30, figsize=(10,5))
plt.suptitle("Distribution of Weight, BMI, and Fat")
plt.show()
```

## Activity vs Weight

```python
# Checking if more steps or activity correlate with weight/BMI:

activity_cols = ['StepTotal','VeryActiveMinutes_x','FairlyActiveMinutes_x','LightlyActiveMinutes_x','SedentaryMinutes_x']

# Correlation with weight
corr_weight = df[activity_cols + ['WeightKg','BMI','Fat']].corr()[['WeightKg','BMI','Fat']]
print("Correlation of Activity Metrics with Weight/BMI/Fat:")
print(corr_weight)
```

## Scatter Plots for Insights

```python
# Steps vs Weight:

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='StepTotal', y='WeightKg')
plt.title("Total Steps vs Weight")
plt.xlabel("Total Steps")
plt.ylabel("Weight (Kg)")
plt.show()
```

## Calories vs BMI:

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Calories_x', y='BMI')
plt.title("Calories Burned vs BMI")
plt.xlabel("Calories")
plt.ylabel("BMI")
plt.show()
```

## Weight Categories

```python
def weight_category(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif bmi < 25:
        return 'Normal'
    elif bmi < 30:
        return 'Overweight'
    else:
        return 'Obese'

df['WeightCategory'] = df['BMI'].apply(weight_category)

# Count plot
sns.countplot(data=df, x='WeightCategory', order=['Underweight','Normal','Overweight','Obese'])
plt.title("Distribution of Users by Weight Category")
plt.show()
```

## Distance vs Weight 

```python
# Checking if users with more active distance have lower weight:

distance_cols = ['VeryActiveDistance_x','ModeratelyActiveDistance_x','LightActiveDistance_x','SedentaryActiveDistance_x']

corr_distance_weight = df[distance_cols + ['WeightKg']].corr()['WeightKg']
print("Correlation between Distance and Weight:")
print(corr_distance_weight)
```

## Weight vs Sedentary Minutes

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='SedentaryMinutes_x', y='WeightKg')
plt.title("Sedentary Minutes vs Weight")
plt.xlabel("Sedentary Minutes")
plt.ylabel("Weight (Kg)")
plt.show()

# Optional correlation value
print("Correlation:", df[['SedentaryMinutes_x','WeightKg']].corr().iloc[0,1])
```

## BMI vs Very Active Minutes (Boxplot)

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='WeightCategory', y='VeryActiveMinutes_x',
            order=['Underweight','Normal','Overweight','Obese'])
plt.title("Very Active Minutes vs Weight Category")
plt.xlabel("Weight Category")
plt.ylabel("Very Active Minutes")
plt.show()
```
