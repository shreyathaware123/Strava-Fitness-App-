#### Distribution of steps

# Calories Analysis

```python
# Using only one Calories column (Calories_x)
calories = df['Calories_x']

#  Average calories burned
print("Average daily calories burned:", calories.mean())

#  Maximum calories burned
print("Maximum calories burned in a day:", calories.max())

#  Minimum calories burned
print("Minimum calories burned in a day:", calories.min())

#  Distribution plot
plt.figure(figsize=(8,5))
sns.histplot(calories, bins=30, color='orange')
plt.title("Distribution of Daily Calories Burned")
plt.xlabel("Calories")
plt.ylabel("Number of Days")
plt.show()
```

# Active Minutes Analysis

```python
#  Average active minutes
avg_active = df[['VeryActiveMinutes_x', 'FairlyActiveMinutes_x', 
                           'LightlyActiveMinutes_x', 'SedentaryMinutes_x']].mean()

print("Average Active Minutes:")
print(avg_active)

#  Distribution plot
df[['VeryActiveMinutes_x', 'FairlyActiveMinutes_x', 
              'LightlyActiveMinutes_x', 'SedentaryMinutes_x']].hist(bins=20, figsize=(10,6))

plt.suptitle("Distribution of Active Minutes")
plt.show()
```

# Total Distance Analysis

```python
import matplotlib.pyplot as plt

# Statistical summary of distances
distance_stats = df[['VeryActiveDistance_x', 'ModeratelyActiveDistance_x', 
                     'LightActiveDistance_x', 'SedentaryActiveDistance_x', 
                     'TotalDistance']].describe()

print("Distance Summary:")
print(distance_stats)

# Average distance by activity type (bar plot)
avg_distance = df[['VeryActiveDistance_x', 'ModeratelyActiveDistance_x', 
                   'LightActiveDistance_x', 'SedentaryActiveDistance_x']].mean()

avg_distance.plot(kind='bar', figsize=(8,5))
plt.title("Average Distance by Activity Type")
plt.ylabel("Distance (km)")
plt.show()
```

# Correlation Between Steps and Calories

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Correlation between Steps and Calories
corr = df[['StepTotal', 'Calories_x']].corr()
print("Correlation between Steps and Calories:")
print(corr)

# Scatter plot
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='StepTotal', y='Calories_x')
plt.title("Steps vs Calories Burned")
plt.xlabel("Total Steps")
plt.ylabel("Calories Burned")
plt.show()
```
