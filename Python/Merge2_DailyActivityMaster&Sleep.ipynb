```python
import pandas as pd
```

```python
df1 = pd.read_csv("Merge_daily_activity_master.csv")
```

```python
df2 = pd.read_csv("clean_sleepDay.csv")
```

```python
df = pd.merge(df1, df2, 
                          left_on=['Id','ActivityDay'], 
                          right_on=['Id','SleepDay'],
                          how='inner')

```

```python
df.head()
```

# Now we can analyze: Steps vs Sleep, Active Minutes vs Sleep

# Save the merge dataset

```python
df.to_csv("Merge_DailyActivity&Sleep.csv", index = False)
```
