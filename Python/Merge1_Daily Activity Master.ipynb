## Standardize the date column in all datasets

```python
import pandas as pd

# Load datasets
activity = pd.read_csv("clean_dailyActivity.csv")
calories = pd.read_csv("clean_dailyCalories.csv")
intensity = pd.read_csv("clean_dailyIntensities.csv")
steps = pd.read_csv("clean_dailySteps.csv")

# Standardize the date column name in all datasets
activity.rename(columns={'ActivityDate':'ActivityDay'}, inplace=True)
calories.rename(columns={'ActivityDay':'ActivityDay'}, inplace=True)
intensity.rename(columns={'ActivityDay':'ActivityDay'}, inplace=True)
steps.rename(columns={'ActivityDay':'ActivityDay'}, inplace=True)
```

## Merge Daily Datasets

```python
daily_master = activity.merge(calories, on=['Id','ActivityDay'], how='inner') \
                       .merge(intensity, on=['Id','ActivityDay'], how='inner') \
                       .merge(steps, on=['Id','ActivityDay'], how='inner')
```
