### Convert ActivityHour into Date + Time Column

```python
df['ActivityHour'] = pd.to_datetime(df['ActivityHour'], format='mixed')
```

### Extract Date (dd/mm/yyyy)

```python
df['Date'] = df['ActivityHour'].dt.strftime('%d/%m/%Y')
```

```python
df['Date']
```

### Extract Hour

```python
df['Hour'] = df['ActivityHour'].dt.hour
```
