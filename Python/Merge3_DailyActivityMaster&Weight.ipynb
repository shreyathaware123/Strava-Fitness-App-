```python
import pandas as pd
```

```python
df1 = pd.read_csv("Merge_daily_activity_master.csv")
```

```python
df2 = pd.read_csv("clean_weightLogInfo.csv")
```

```python
df = pd.merge(df1, df2, 
                           left_on=['Id','ActivityDay'], 
                           right_on=['Id','Date'],
                           how='inner')
```

```python
df.head()
```

# Now we can analyze: Steps vs Weight, Calories vs Weight, BMI trends

# save the merge dataset

```python
df.to_csv("Merge_DailyActivity&Weight.csv", index = False)
```
