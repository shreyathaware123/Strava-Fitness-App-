# ** Eda Insights **

## Sleep Overview

```python
# How much people are actually sleeping:

print("Average Minutes Asleep:", df['TotalMinutesAsleep'].mean())
print("Average Time in Bed:", df['TotalTimeInBed'].mean())
print("Number of Sleep Records per User (summary):")
print(df['TotalSleepRecords'].value_counts())
```

## Distribution plots:

```python
plt.figure(figsize=(8,5))
sns.histplot(df['TotalMinutesAsleep'], bins=30, color='blue')
plt.title("Distribution of Total Minutes Asleep")
plt.xlabel("Minutes Asleep")
plt.ylabel("Number of Days")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['TotalTimeInBed'], bins=30, color='green')
plt.title("Distribution of Total Time in Bed")
plt.xlabel("Time in Bed (minutes)")
plt.ylabel("Number of Days")
plt.show()
```

## Steps vs Sleep

```python
# Does more activity result in longer sleep?

print("Correlation between Steps and Sleep:")
print(df[['StepTotal','TotalMinutesAsleep']].corr())

plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='StepTotal', y='TotalMinutesAsleep')
plt.title("Steps vs Sleep Duration")
plt.xlabel("Total Steps")
plt.ylabel("Total Minutes Asleep")
plt.show()
```

## Active Minutes vs Sleep

```python
# Checking which type of activity affects sleep most?:

active_cols = ['VeryActiveMinutes_x','FairlyActiveMinutes_x','LightlyActiveMinutes_x','SedentaryMinutes_x']

active_sleep_corr = df[active_cols + ['TotalMinutesAsleep']].corr()['TotalMinutesAsleep']
print("Correlation of Active Minutes with Sleep:")
print(active_sleep_corr)
```

## Sleep Efficiency

```python
# We can create a new column: how efficient a user is sleeping:

df['SleepEfficiency'] = df['TotalMinutesAsleep'] / df['TotalTimeInBed'] * 100

print("Average Sleep Efficiency:", df['SleepEfficiency'].mean())
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['SleepEfficiency'], bins=30, color='purple')
plt.title("Distribution of Sleep Efficiency (%)")
plt.xlabel("Sleep Efficiency (%)")
plt.ylabel("Number of Days")
plt.show()
```

```python
# Helps identify if users spend too much time in bed but don’t sleep well.
```

## Sleep vs Day of Week Patterns

```python
# People’s sleep may vary by day of the week (weekdays vs weekends).
```

```python
df['ActivityDay'] = pd.to_datetime(df['ActivityDay'], dayfirst=True)
df['DayOfWeek'] = df['ActivityDay'].dt.day_name()

sleep_by_day = df.groupby('DayOfWeek')['TotalMinutesAsleep'].mean().reindex(
    ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'])

print("Average Sleep by Day of the Week:")
print(sleep_by_day)

# Plot
sleep_by_day.plot(kind='bar', figsize=(8,5), color='skyblue')
plt.title("Average Sleep by Day of the Week")
plt.ylabel("Total Minutes Asleep")
plt.show()
```

## Sleep vs Very Active Minutes

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='VeryActiveMinutes_x', y='TotalMinutesAsleep')
plt.title("Very Active Minutes vs Sleep Duration")
plt.xlabel("Very Active Minutes")
plt.ylabel("Total Minutes Asleep")
plt.show()

# Optional: correlation value
print("Correlation:", df[['VeryActiveMinutes_x','TotalMinutesAsleep']].corr().iloc[0,1])
```

## Sleep Consistency

```python
user_avg_sleep = df.groupby('Id')['TotalMinutesAsleep'].mean().reset_index()
user_avg_sleep.columns = ['Id', 'AvgSleep']

df = df.merge(user_avg_sleep, on='Id', how='left')
df['SleepDeviation'] = df['TotalMinutesAsleep'] - df['AvgSleep']

plt.figure(figsize=(8,5))
sns.histplot(df['SleepDeviation'], bins=30, color='coral')
plt.title("Distribution of Sleep Deviation from User Average")
plt.xlabel("Sleep Deviation (minutes)")
plt.ylabel("Number of Records")
plt.show()
```
